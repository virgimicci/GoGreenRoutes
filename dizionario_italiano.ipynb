```python
import pandas as pd
import json
import numpy as np
```

```python
df = pd.read_excel('NRC-Emotion-Lexicon-v0.92/NRC-Emotion-Lexicon-v0.92-In105Languages-Nov2017Translations.xlsx', index_col=0)  
```

```python
df.head()
```

```python
df = df.reset_index(level='English (en)', col_level=1) 
```

```python
df.columns= df.columns.str.strip().str.lower()
df.columns
```

```python
df.head()
```

Qua seleziono la lingua!!

```python
lingua = 'italian (it)'
```

```python
df_s = df[[lingua,'positive', 'negative', 'anger', 'anticipation', 'disgust', 'fear',
       'joy', 'sadness', 'surprise', 'trust']]
df_s.head()
```

```python
def checkSentiment(x):
    columns = ['positive', 'negative', 'anger', 'anticipation', 'disgust', 'fear',
       'joy', 'sadness', 'surprise', 'trust']
    sentiments = []
    for col in columns:
       if (x[col] == 1):
        sentiments.append(col)  
    return sentiments
```

```python
df_s['sentiments'] = df_s.apply(checkSentiment, axis=1)
df_s
```

```python
df_l=df_s[[lingua,'sentiments']]
df_l
```

Cambio nome alla colonna

```python
abbr = 'it' 
df_l = df_l.rename(columns={lingua : abbr})
df_l.columns
```

Elimino le liste vuote

```python
df_l.sentiments = df_l.sentiments.apply(lambda y: np.nan if len(y)==0 else y)
df_l.it = df_l.it.apply(lambda y: np.nan if y=='NO TRANSLATION' else y)
df_l.head()
```

```python
df_l.dropna(subset=['sentiments'], inplace=True)
df_l.head()
```

```python
a = df_l.loc[df_l['it'] == 'compatibile']
a
```

```python
print(df_l.loc[[2415]])
```

```python
dict = pd.Series(df_l.sentiments.values,index=df_l.it).to_dict()
```

```python
with open('nrc_ita.json', 'w') as fp:
    json.dump(dict, fp,  indent=4)
```
